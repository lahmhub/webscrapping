```python
import requests
import pandas as pd
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
```

```python
ua = UserAgent()
userAgent = ua.random
headers = {'User-Agent': userAgent}
```

```python
url = 'https://www.homedepot.com/b/Appliances-Washers-Dryers-Washing-Machines-Front-Load-Washers/N-5yc1vZc3pj'
```

```python
headers = {'User-Agent': userAgent}
```

```python
page = requests.get(url, headers = headers)
```

```python
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
flws=soup.find("div", attrs={"data-component": "product-results:ResultsWrapped:v13.6.6"})
names=[]
ratings=[]
pricing=[]
for flw in flws:
    for flw_names in flw.find_all("span", attrs={"data-testid": 'attribute-product-label'}):
        #name
        if flw_names is not None:
            names.append(flw_names.text)
        else:
            names.append('None')

    
    for flws_ratings in flw.find_all("a", attrs={"data-testid": 'product-pod__ratings-link'}):
        if flws_ratings is not None:
            ratings.append(flws_ratings.text)
        else:
            ratings.append('None')
    

    
    for flws_pricing in flw.find_all("div", attrs={"data-component": "price:Price:v6.31.5"}):
        if flws_pricing is not None:
            pricing.append(flws_pricing.text)
        else:
            pricing.append('None')
    
print(names)
print(ratings)
print(pricing)
```

```python
import pandas as pd
washer_df=pd.DataFrame({
    "product":names,
    "ratings":ratings,
    "pricing":pricing
})
```

```python
washer_df
```

```python
washer_df["pricing"].str.replace("Limit "," ").str.replace("$","|").str.replace(".00Was","").str.replace(".00 Save ","").str.replace(".00 ","")
```

```python
washer_df["pricing"]=washer_df["pricing"].str.replace("Limit "," ").str.replace("$","|").str.replace(".00Was","").str.replace(".00 Save ","").str.replace(".00 ","")
```

```python
washer_df
```

```python
washer_df[["limit","Discounted Price","Original Price","Discount"]]=washer_df["pricing"].str.split("|", expand=True)
```

```python
washer_df
```

```python
washer_df[["Discount Amount","Discount Percent"]]=washer_df["Discount"].str.split("(",expand=True)
```

```python
washer_df
```

```python
washer_df["Discount Percent"]=washer_df["Discount Percent"].str.replace(".00%)",'').str.replace("Buy More","|Buy More")
```

```python
washer_df[["Discount Percent","to_drop"]]=washer_df["Discount Percent"].str.split("|", expand=True)
```

```python
washer_df.drop(columns=["pricing","Discount","to_drop"], axis=1, inplace=True)
```
